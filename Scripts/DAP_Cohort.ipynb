```python
#Combine cognitive score data and dog meta data . Need only - dog_id,treat_date,estimated_dob,cognitive score columns
dog_data_combined=pd.merge(df_long,ages_df,on='dog_id',how='left')
dog_data_combined
```

```python
#calculate age
# Convert date columns to datetime
dog_data_combined['trt_game_date'] = pd.to_datetime(dog_data_combined['trt_game_date'])
dog_data_combined['Estimated_DOB'] = pd.to_datetime(dog_data_combined['Estimated_DOB'])

# Compute dog age in years
dog_data_combined['age_years'] = (dog_data_combined['trt_game_date'] - dog_data_combined['Estimated_DOB']).dt.days / 365.25
dog_data_combined['age_years']=dog_data_combined['age_years'].astype('int')
dog_data_combined
```

```python
dog_data_combined[dog_data_combined['dog_id']=='273']
```

```python
# Step 1: Remove rows where accuracy is NaN (dog didn't respond)
#dog_data_combined[dog_data_combined['accuracy'].isna()] #n =4918
df_valid = dog_data_combined.dropna(subset=['accuracy'])
```

```python
df_valid
```

```python
# Step 2: Group by dog_id (and optionally age if available)
cognitive_score_df = df_valid.groupby(['dog_id','age_years']).agg(
    prop_correct=('accuracy', 'mean'),  # Mean of 0/1 = proportion correct
    n_trials=('accuracy', 'count')      # Number of valid (non-NaN) trials
).reset_index()
```
