```python
def hellinger_transform(df_counts: pd.DataFrame) -> pd.DataFrame:
    """Row-normalize to relative abundance, then sqrt (Hellinger)."""
    rel = df_counts.div(df_counts.sum(axis=1).replace(0, np.nan), axis=0).fillna(0.0)
    return np.sqrt(rel)
```

```python
from sklearn.decomposition import PCA
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def run_pca(X: pd.DataFrame, n_components: int) -> tuple[pd.DataFrame, pd.DataFrame, PCA]:
    """Fit PCA and return (scores, loadings, pca_model)."""
    n_components = min(n_components, X.shape[1])
    pca = PCA(n_components=n_components)
    scores = pca.fit_transform(X.values)
    scores_df = pd.DataFrame(scores, index=X.index,
                             columns=[f"PC{i+1}" for i in range(pca.n_components_)])
    loadings_df = pd.DataFrame(pca.components_.T, index=X.columns,
                               columns=[f"PC{i+1}" for i in range(pca.n_components_)])
    return scores_df, loadings_df, pca
```

```python
def correlate_pcs_with_trait(scores_df: pd.DataFrame, trait: pd.Series) -> pd.DataFrame:
    rows = []
    for pc in scores_df.columns:
        r_pea, p_pea = pearsonr(scores_df[pc], trait)
        r_spr, p_spr = spearmanr(scores_df[pc], trait)
        rows.append({
            "PC": pc,
            "Pearson_r": r_pea, "Pearson_p": p_pea,
            "Spearman_rho": r_spr, "Spearman_p": p_spr
        })
    return pd.DataFrame(rows)
```

```python
def top_loading_species(loadings_df: pd.DataFrame, k=10) -> dict:
    """Return dict: PC -> DataFrame of top + and top - species by loading magnitude."""
    out = {}
    for pc in loadings_df.columns:
        s = loadings_df[pc].sort_values(ascending=False)
        top_pos = s.head(k).to_frame(name="loading")
        top_neg = s.tail(k).to_frame(name="loading")
        out[pc] = (top_pos, top_neg)
    return out
```

```python
def quick_barplot_variance(pca: PCA, title: str, path: str):
    import matplotlib.pyplot as plt
    import numpy as np

    # Variance explained by each PC
    var_exp = pca.explained_variance_ratio_ * 100.0
    pcs = np.arange(1, pca.n_components_ + 1)

    plt.figure(figsize=(10, 6))
    bars = plt.bar(pcs, var_exp)

    # Add labels above each bar
    for bar, val in zip(bars, var_exp):
        plt.text(
            bar.get_x() + bar.get_width() / 2,   # x position (center of bar)
            bar.get_height() + 0.5,              # y position slightly above bar
            f"{val:.1f}%",                       # format as %
            ha="center", va="bottom", fontsize=8, rotation=90
        )

    plt.xlabel("Principal Component")
    plt.ylabel("% Variance Explained")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(path, dpi=150)
    plt.close()
```

```python
def quick_scatter(scores_df: pd.DataFrame, trait: pd.Series, title: str, path: str,
                  x="PC1", y="PC2"):
    plt.figure()
    sc = plt.scatter(scores_df[x], scores_df[y], c=trait.values)
    plt.colorbar(sc, label=trait.name if trait.name else "Trait")
    plt.xlabel(x); plt.ylabel(y); plt.title(title)
    plt.tight_layout()
    plt.savefig(path, dpi=150)
    plt.close()
```

```python
# === CLEAN ALIGNMENT OF ABUNDANCE & COG TABLES ===
def norm_id(x: str) -> str:
    """Normalize sample IDs so they match across files."""
    x = str(x).strip()
    x = x.split('/')[-1].split('\\')[-1]  # drop paths
    x = x.split('.')[0]                   # drop extensions (e.g., .fastq.gz)
    return x

def require_cols(df: pd.DataFrame, needed: list, dfname: str = "DataFrame"):
    missing = [c for c in needed if c not in df.columns]
    if missing:
        raise ValueError(f"{dfname} is missing columns: {missing}. "
                         f"Has: {list(df.columns)[:10]}...")
```

```python
# 0) Expect: abundance_df has a column 'SRA_ID' + species columns.
#            cog has columns 'SRA_ID' and 'CognitiveScore'.
require_cols(abundance_df, ["SRA_ID"], "abundance_df")
require_cols(cog, ["SRA_ID", "CognitiveScore"], "cog")
```

```python
# 1) Normalize IDs on both sides
abundance_df = abundance_df.copy()
cog = cog.copy()
abundance_df["SRA_ID"] = abundance_df["SRA_ID"].map(norm_id)
cog["SRA_ID"]          = cog["SRA_ID"].map(norm_id)
```

```python
# 2) Drop obvious bad rows / duplicates, then set index
abundance_df = (abundance_df
                .dropna(subset=["SRA_ID"])
                .drop_duplicates(subset=["SRA_ID"])
                .set_index("SRA_ID"))
```

```python
cog = (cog
       .dropna(subset=["SRA_ID", "CognitiveScore"])
       .drop_duplicates(subset=["SRA_ID"])
       .set_index("SRA_ID"))
```

```python
# 3) Align by overlapping SRA_IDs
common = abundance_df.index.intersection(cog.index)
print("overlap n =", len(common))
if len(common) == 0:
    # helpful diagnostics
    print("Example IDs (abundance):", list(abundance_df.index[:5]))
    print("Example IDs (cog):", list(cog.index[:5]))
    raise ValueError("No overlapping SRA_IDs after normalization. "
                     "Inspect ID formats and the norm_id() function.")

abundance_df = abundance_df.loc[common].sort_index()
trait = cog.loc[common, "CognitiveScore"].astype(float)

print("aligned abundance_df shape:", abundance_df.shape)
print("aligned trait shape:", trait.shape)
```

```python
# 4) Optional pruning: drop species that are zero in all aligned samples
nonzero_cols = (abundance_df > 0).any(axis=0)
abundance_df = abundance_df.loc[:, nonzero_cols]
print("columns after dropping all-zero species:", abundance_df.shape[1])
```

```python
# === ready for tb-PCA (Hellinger -> PCA) ===
def hellinger_transform(df_counts: pd.DataFrame) -> pd.DataFrame:
    rel = df_counts.div(df_counts.sum(axis=1).replace(0, np.nan), axis=0).fillna(0.0)
    return np.sqrt(rel)
```

```python
def plot_cumulative_variance(pca, title, path):
    import numpy as np
    import matplotlib.pyplot as plt

    # cumulative variance from the argument `pca`, not hardcoded
    cumvar = np.cumsum(pca.explained_variance_ratio_) * 100

    plt.figure(figsize=(6,4))
    plt.plot(range(1, len(cumvar)+1), cumvar, marker='o')
    plt.axhline(90, color='r', linestyle='--', label="90% threshold")
    plt.xlabel("Principal Component")
    plt.ylabel("Cumulative Variance Explained (%)")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.savefig(path, dpi=150)
    plt.close()

    # PCs needed for ~90%
    n_pcs_90 = np.argmax(cumvar >= 90) + 1 if np.any(cumvar >= 90) else None
    if n_pcs_90:
        print(f"≈90% of variance is explained by the first {n_pcs_90} PCs.")
    else:
        print("Even with all PCs, variance explained is <90%.")
plot_cumulative_variance(
    pca_hel,  # or pca_clr
    "Cumulative Variance Explained (Hellinger)", 
    os.path.join(OUTPUT_DIR, "cumvar_tbPCA.png")
)
```

```python
#Fit regression line
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress  # if you don't have SciPy, see the polyfit fallback below

# x and y
x = scores_hel["PC3"].to_numpy()
y = np.asarray(trait.values, dtype=float)

# drop NaNs just in case
mask = np.isfinite(x) & np.isfinite(y)
x, y = x[mask], y[mask]

# fit y = m*x + b
res = linregress(x, y)  # res.slope, res.intercept, res.rvalue, res.pvalue
x_line = np.linspace(x.min(), x.max(), 200)
y_line = res.slope * x_line + res.intercept

# plot
plt.figure()
sc = plt.scatter(x, y, c=y, cmap="viridis")
plt.plot(x_line, y_line, linewidth=2, color="black",
         label=f'y = {res.slope:.3f}x + {res.intercept:.3f} '
               f'(R²={res.rvalue**2:.3f}, p={res.pvalue:.2e})')
plt.legend()

plt.colorbar(sc, label=trait.name if trait.name else "CognitiveScore")
plt.xlabel("PC3")
plt.ylabel("CognitiveScore")
plt.title("PC3 vs CognitiveScore")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "tbPCA_PC3_cognition_with_regline.png"), dpi=150)
plt.close()
```
